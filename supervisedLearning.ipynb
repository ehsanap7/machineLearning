# KNN

```python
def confusion_matrix_scorer(clf, X, y):
    y_pred = clf.predict(X)
    cm = confusion_matrix(y, y_pred)
    return {'TN': cm[0, 0], 'FP': cm[0, 1],'FN': cm[1, 0], 'TP': cm[1, 1]
             ,'PPV':np.divide(cm[1, 1], cm[1, 1]+cm[0, 1],where=cm[1, 1]+cm[0, 1]!=0)
             ,'NPV':np.divide(cm[0, 0], cm[0, 0]+cm[1, 0],where=cm[0, 0]+cm[1, 0]!=0)
             ,'Sensitivity':np.divide(cm[1, 1], cm[1, 1]+cm[1, 0],where=cm[1, 1]+cm[1, 0]!=0)
             ,'Specificity':np.divide(cm[0, 0], cm[0, 0]+cm[0, 1],where=cm[0, 0]+cm[0, 1]!=0)
             ,'Accuracy':np.divide(cm[1, 1]+cm[0,0], cm[0, 0]+cm[0, 1]+cm[1, 0]+cm[1, 1],where=cm[0, 0]+cm[0, 1]+cm[1, 0]+cm[1, 1]!=0)}
```

```python
Ks_circle = 21
listframe_circle = [] 
mean_acc_circle = np.zeros((Ks_circle-1))
k_scores_circle = []

for n in range(1,Ks_circle):
    neigh_circle = KNeighborsClassifier(n_neighbors = n)
    cv_results_circle = cross_validate(neigh_circle, X_circle, y_circle, cv=10,scoring=confusion_matrix_scorer)
    dataFrame_circle=pd.DataFrame(cv_results_circle)
    meanValue_circle = dataFrame_circle.mean()['test_Accuracy']
    k_scores_circle.append(meanValue_circle)
    listframe_circle.append(dataFrame_circle) 
```

```python
plt.plot(range(1,Ks_circle), k_scores_circle)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.xticks(range(0,21,1))
# plt.yticks(np.arange(0.9, 1.2, step=0.09)) 
# plt.gcf().set_size_inches(10, 7)
plt.title("Zahra KH, Lida M, Ehsan Ap",color='red')
plt.show()
```

```python
acc_k_circle=max(k_scores_circle)
k_circle=k_scores_circle.index(acc_k_circle)
listframe_circle[k_circle]
print( "The best accuracy was with", acc_k_circle, "with k=", k_circle+1) 
```

```python
print(np.mean(listframe_circle[k_circle]['test_PPV']))
print(np.mean(listframe_circle[k_circle]['test_NPV']))
print(np.mean(listframe_circle[k_circle]['test_Sensitivity']))
print(np.mean(listframe_circle[k_circle]['test_Specificity']))
print(np.mean(listframe_circle[k_circle]['test_Accuracy']))
```
